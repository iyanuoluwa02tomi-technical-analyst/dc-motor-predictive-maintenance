# PROJECT OVERVIEW:
This project implements a Real time condition Monitoring sysem designed to reduce industrial downtime. By analysing Vibration and thermal data ,The system identifies mechanical irregularities.
#TECHNICAL IMPLEMENTATION
*signal Processing: Utilized Numpy for RMS calculations.
*Data simulation: Modeled Motor behaviour using 'pandas' and 'Matplotlib' 
*predictive Logic: custom 'Motor' class with automated diagnostic thresholds

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt

time=np.linspace(0,5,1000)
healthy_signal=0.5*np.sin(2*np.pi*10*time)
noise=np.random.normal(0,0.2,1000)
faulty_signal=healthy_signal+noise

faulty_signal[600:620]
faulty_signal[600:620]+1.5

data=pd.DataFrame({'Time_sec':time,'Healthy_vibration':healthy_signal,'Faulty_vibration':faulty_signal})
print("Dataset preview")
print(data.head())
```

```python
plt.figure(figsize=(12,6))
plt.plot(data['Time_sec'],data['Healthy_vibration'],'r',label='Healthy motor',alpha=1.0)
plt.plot(data['Time_sec'],data['Faulty_vibration'],'b',label='Faulty Motor',alpha=0.6)
plt.xlabel('Time_sec(s)')
plt.ylabel('Amplitude')
plt.title('Predictive Maintenance Analysis')
plt.legend(loc=0)
plt.show()
```

```python
Square_of_Faulty_Vibration=data['Faulty_vibration']**2
Square_of_Faulty_vibration_mean=np.mean(Square_of_Faulty_Vibration)
rms_of_Faulty_vibration=np.sqrt(Square_of_Faulty_vibration_mean)

Square_of_Healthy_Vibration=data['Healthy_vibration']**2
Square_of_Healthy_vibration_mean=np.mean(Square_of_Healthy_Vibration)
rms_of_Healthy_vibration=np.sqrt(Square_of_Healthy_vibration_mean)
```

```python
print(rms_of_Faulty_vibration)
print(rms_of_Healthy_vibration)
```

```python
# Decision Logic
current_rms=rms_of_Faulty_vibration
Baseline=rms_of_Healthy_vibration
if current_rms>Baseline+(Baseline*10/100 ):
     print('WARNING')
else:
     print('GOOD CONDITION')
     
```

```python
class Motor:
    def __init__(self,name,vibration_data,baseline,temperature):
        self.name=name
        self.data=vibration_data
        self.baseline=baseline
        self.temperature=temperature

    def calculations_for_rms(self):
        rms_value=np.sqrt(np.mean(self.data**2))
        return rms_value

    def check_health(self):
        current_rms=self.calculations_for_rms()
        if  current_rms>self.baseline+(self.baseline*10/100 ) or (self.temperature>70):
             return 'WARNING'
        else:
             return 'GOOD CONDITION'
            
    def get_diagnostic(self,current_temp):
        rms_value=self.calculations_for_rms()
        threshold=self.baseline*1.10
        if rms_value>threshold and current_temp>70:
            return "EMERGENCY"
        elif rms_value>threshold or current_temp>70:
            return "WARNING"
        else:
            return "NORMAL"
            
motor1=Motor("conveyor_belt",np.array([0.1,0.2]),0.35,49)
motor2=Motor("cooling_fan",np.array([0.05,0.08]),0.35,69)
motor3=Motor("broken_pump",np.array([0.8,0.9]),0.35,82)

print(motor1.check_health())
print(motor2.check_health())
print(motor3.check_health())
```
